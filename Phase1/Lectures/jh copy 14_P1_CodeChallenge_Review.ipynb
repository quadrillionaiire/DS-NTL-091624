## Task 14: Create a new dictionary

Create a dictionary called `song_dictionary` which consists of each track name `string` as a key and a `tuple` of artists associated with each track as a value.

```python
# call_on_students(1)
```

```python
# Your code here:
# Create a dictionary with track names as keys and artists as values
song_dictionary = {item['name']: tuple(artist['name'] for artist in item['artists']) for item in first_twenty_songs}
```

In this case, the keys will be the names of songs, and the values will be the names of the artists who created those songs.

Breakdown
song_dictionary = {...}: This part indicates that we are creating a new dictionary and storing it in a variable named song_dictionary.

{item['name']: ... for item in first_twenty_songs}: This is a way of saying, "For each item in the list first_twenty_songs, use item['name'] (which is the song name) as the key." The for loop goes through each song in the first_twenty_songs list.

item['name']: Here, item represents a single song from the list. item['name'] gets the name of that song. So if the song is "Song A," this would return "Song A."

tuple(artist['name'] for artist in item['artists']): This part creates the value for the dictionary. Itâ€™s saying, "For each artist in item['artists'] (which is a list of artists for that song), get their names and put them into a tuple." A tuple is similar to a list but canâ€™t be changed after it's created.

artist['name']: This gets the name of an individual artist from the list of artists for that song.

## Task 15: Write a function

Create a function with takes an **artist name** and the **song_dictionary** as arguments, and returns a `list` of songs written by that artist. 

```python
# call_on_students(1)
```

```python
# Easier to do things outside of a function first
# Let's try for The Weeknd
```

```python
# Your code here

def find_song_by_artist(artist_name, song_dict):
    
    '''
    Parameters:
    arist_name: a string of an artist's name to be used to search the dictionary
    song_dict:  a dictionary of top_twenty songs with song name as keys and a list of 
    artist names as values
    
    Returns:
    A list of songs which the given artist appeared on
    '''
    
    songs_by_artist = [song for song, artists in song_dict.items() if artist_name in artists]
    return songs_by_artist

# Test the function
print(find_song_by_artist('The Weeknd', song_dictionary))
```

songs_by_artist = [song for song, artists in song_dict.items() if artist_name in artists]: This line is where the magic happens. 

It creates a list called songs_by_artist. Hereâ€™s how it works:

song_dict.items(): This gets all the key-value pairs from the dictionary, where song is the song name and artists is the list of artists for that song.

if artist_name in artists: This checks if the given artist_name is in the list of artists for each song. If it is, that song gets added to our list.
Returning the Result:

return songs_by_artist: This line sends back the list of songs that include the specified artist. If the artist isnâ€™t found, the list will just be empty.
